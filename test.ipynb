```python
import rasterio
import matplotlib.pyplot as plt

# Replace with the path to your TIFF file
tif_path = "/mnt/storage/code/EOV_NDVI/brios/datasets/TRAINING_Data4BRIOS_4M/BinhTrung-BinhSon-QuangNgai/binhtrung_rvi_8days/binhtrung_rvi_8days_2022-10-01.tif"

# Open the TIFF file using rasterio
with rasterio.open(tif_path) as src:
    bands = src.count  # Get the number of bands

    # Create subplots: one subplot per band
    fig, axs = plt.subplots(1, bands, figsize=(5 * bands, 5))
    
    # If there is only one band, wrap axs in a list for consistency
    if bands == 1:
        axs = [axs]
    
    # Loop through each band and display it
    for i in range(bands):
        band_data = src.read(i + 1)  # Bands are 1-indexed in rasterio
        axs[i].imshow(band_data, cmap='gray')
        axs[i].set_title(f'Band {i+1}')
        axs[i].axis('off')

    plt.tight_layout()
    plt.show()
```
